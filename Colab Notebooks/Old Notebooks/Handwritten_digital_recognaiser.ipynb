```python
# Importing modules
from PIL import Image
import mnist
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
```

```python
# creating training variabls
X_train=mnist.train_images() 
y_train=mnist.train_labels()
X_test=mnist.train_images()
y_test=mnist.train_labels()
```

```python

print("X-train",X_train)
print("Y-train",y_train)
print("X-test",X_test)
print("Y-test",y_test)
```

```python
# To know the dimntions of array
X_train.ndim
```

```python
X_train.shape
# 28,28 are the pixles and 60,000 are number of them
```

```python
# Instead putting 60,000 manually python will auto detect numbers by putting -1
X_train=X_train.reshape((-1,28*28))
X_test=X_test.reshape((-1,28*28))
```

```python
# Printing the d=first element in X_train
print("X_train",X_train[0])
```

```python
X_train=(X_train/256)
X_test=(X_test/256)
```

```python
print(X_train[0])
```

```python
clf=MLPClassifier(solver='adam', activation='relu', hidden_layer_sizes=(64,64))
```

```python
clf.fit(X_train,y_train)
```

```python
prediction=clf.predict(X_test)
```

```python
# accuracy variable( acc )
acc=confusion_matrix(y_test,prediction)
print(acc)
```

```python
# Creating a variable to calculate accuracy 
# cm=confusion_matrix
def accuracy(cm):
  # making a diagonal variable to sum of diagonals
  diagonal=cm.trace()
  elements=cm.sum()
  return diagonal/elements
```

```python
print(accuracy(acc))
```
