```python
num=[]
import re
vars =['a1c2b','21ang','92c','451c',2]
for i in vars:
  num.append(re.findall(r'\d+',str(i)))

no=[]
for i in num:
  if len(i)>0:
    for j in i:
        no.append(j)
  else:
    no.append(i)

sum=[]
for i in no:
  if len(i)>0:
    for j in i:
        sum.append(int(j))
  else:
    sum.append(int(i))


res=0
for i in sum:
  res=res+i

print(res)
```
