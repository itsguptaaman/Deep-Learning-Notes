$$NN = f(weights, biases)$$ 

```python
import torch
```

```python
# To calculate gradient with respect to x we need to do requires_grad=True
# x = torch.tensor(2.0, requires_grad=False)
x = torch.tensor(2.0, requires_grad=True)

# Example
f = x ** 2 + x ** 3
print(f)
```

```python
f.backward()
```

```python
x.grad
```

```python
x.grad.item()
```

```python
# Partial Derivative
x = torch.tensor(2.0, requires_grad=True)
y = torch.tensor(1.0, requires_grad=True)

f = x ** 2 + 2 * y ** 3

print(f)
```

```python
x.grad
```

```python
#  df wrt x (df/dx) and df wrt y (df/dy) 
f.backward()
```

```python
f.backward() # df/
```

```python
x.grad, y.grad
```

```python
x.grad.item(), y.grad.item()
```
